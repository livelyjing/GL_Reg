# BA Graphs

```python
G_BA = nx.barabasi_albert_graph(20,1,seed = 777)
L_BA = nx.laplacian_matrix(G_BA).toarray()
normL_BA = (L_BA.shape[0]/np.trace(L_BA))*L_BA
```

```python
Signals, Regressors = glr.genAllDist(G_BA)
```

```python
Results = [glr.findHParams(Signals[i], Regressors[i], normL_BA) for i in range(len(Regressors))]
```

```python
for r in Results:
    print(r[0], r[1], round(glr.metricsprf(normL_BA, r[2])[2], 4))
```

```python
L = glr.GSPRegression(Signals[0], Regressors[0], 0.04999999999999993, 1.4, 5, 10**(-3))[0]
glr.metricsprf(normL_BA, L)
```

```python
L = glr.GSPRegression(Signals[2], Regressors[2], 0.009999999999999995, 0.40999999999999986, 5, 10**(-3))[0]
glr.metricsprf(normL_BA, L)
```
